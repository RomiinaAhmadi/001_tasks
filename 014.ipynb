```python
import pandas as pd

df = pd.read_csv('data/heart.csv')


numeric_cols = [
    'age',
    'trestbps',
    'chol',
    'thalach',
    'oldpeak'
]

numeric_df = df[numeric_cols]


stats = pd.DataFrame({
    'Variance': numeric_df.var(),
    'Std': numeric_df.std()
})

stats
```

Age -> مقدار انحراف معیار و واریانس تقریبا متوسط است که این یعنی پراکندگی در سن افراد متوسط بوده یعنی خیلی اختلاف شدیدی در سن افراد نیست پس پیدا کردن الگو و فهمیدن وجود بیماری سخت خواهد بود  

Trestbps ->  مقدار انحراف معیار و واریانس نسبتاً زیاده پس یعنی پراکندگی زیاد بوده پس یعنی مثلا افراد یا فشار خون نرمال داشتن یا فشار خونشون نرمال بوده پس این ویژگی به درد ما می خوره چون پیدا کردن الگو راحت تره 

Chol -> مقدار انحراف معیار و واریانس خیلی بالاست که نشان دهنده پراکندگی زیاد است که میشه گفت ویژگی مهمی برای پیدا کردن بیماری به حساب میاد اما امکان اینکه داده های outlier داشته باشیم وجود داره و شاید همین داده ها باعث بالا بودن انحراف معیار و واریانس بوده 

Thalach -> مقدار انحراف معیار و واریانس نسبتاً بالا هست پست یعنی ضربان قلب افراد متنوع بوده پس می تونه به ما در تشخیص بیماری کمک کند 

Oldpeak ->  مقدار انحراف معیار و واریانس خیلی کم هست  که یعنی پراکندگی خیلی کم است و تشخیص بیماری و پیدا کردن یک الگو سخت خواهد بود
