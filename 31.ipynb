```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.core.pylabtools import figsize

df = pd.read_csv("heart.csv")
corr = df.corr()
corr['target'].sort_values(ascending=False)
sns.heatmap(corr[['target']],
            annot=True,
            cmap="coolwarm")
plt.show()
```
