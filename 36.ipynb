```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df =pd.read_csv('data/heart.csv')
top_features = ['cp','thalach','oldpeak','exang','ca','target']
df_top = df[top_features]

sns.pairplot(df_top, hue= 'target')
plt.show()
```
