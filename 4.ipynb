```python
import pandas as pd

df = pd.read_csv('data/heart.csv')

for column in df.columns:
    count = df [column].nunique()
    if count == 2:
        print(f"{column}: Binary({count} values)")
    elif count <= 10:
        print(f"{column}: Categorical({count}values)")
    else:
        print(f"{column}: Numerical({count} values)")
        
```
