```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
df = pd.read_csv("data/heart.csv")
df
```

### سوال 17

برای تمام ویژگی‌های عددی KDE Plot رسم کنید.

تحلیل:

- نرمال بودن یا نبودن توزیع

```python
df.describe()
```

```python
fig, ax = plt.subplots(nrows=1 ,ncols=2 ,figsize= (15,5))

sns.kdeplot(data = df ,x= "age",fill=True ,color="orange" ,ax = ax[0])
ax[0].set_title("KDE PLOT 'age'")
sns.kdeplot(data = df ,x= "trestbps",fill=True ,color="green",ax = ax[1])
plt.title("KDE PLOT 'trestbps'" )
plt.show()
```

در نمودار سن تقریبا توزیع نرمال داریم ولی نمیشه گفت کاملا نرمال چون یک نقطی از زنگوله خارج شده و نظم بهم زده

در نمودارفشار خون در زمان استراحت توزیع نرمال نداریم چون داده ها در سمت راست بیشتر از سمت چپ هستش و کشیدگی به سمت راست هستش

```python
fig, ax = plt.subplots(nrows=1 ,ncols=3 ,figsize= (20,5))

sns.kdeplot(data = df ,x= "chol",fill=True ,color="pink" ,ax = ax[0])
ax[0].set_title("KDE PLOT 'chol'")
sns.kdeplot(data = df ,x= "thalach",fill=True ,color="skyblue",ax = ax[1])
ax[1].set_title("KDE PLOT 'thalach'" )
sns.kdeplot(data = df ,x= "oldpeak",fill=True ,color="purple",ax = ax[2])
plt.title("KDE PLOT 'oldpeak'" )
plt.show()
```

در نمودار کلسترول و شیب اس تی در تست ورزش مشخصا توزیع نرمال نداریم چون کشیدگی راست داریم

در این یکی نمودار نیز توزیع نرمال نداریم چون این نمودار تالاسمی کشیدگی چپ داریم
