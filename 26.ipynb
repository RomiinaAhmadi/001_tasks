```python
import pandas as pd

df = pd.read_csv('data/heart.csv')


categorical_cols = [
    'sex',
    'cp',
    'fbs',
    'restecg',
    'exang',
    'slope',
    'ca',
    'thal'
]

for col in categorical_cols:

    print(f"\n===== {col} vs target =====")
    
    cross_tab = pd.crosstab(
        df[col],
        df['target'],
        normalize='index'
    ) * 100
    
    print(cross_tab)
```
