```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("heart.csv")
coulms = ['restecg','cp', 'slope', 'thal', 'ca']
fig, axes = plt.subplots(3, 2, figsize=(10, 5))
for ax, col in zip(axes.flatten(), coulms):
    sns.countplot(data=df, x=col, ax=ax)
    ax.set_title(col)
plt.show()
```

از zipاستفاده میکنیم برای اتصال لیست ها
در پرانتز countplotمشخص میکنیم از چه دیتافریمی استفاده کنیم برای برداشتن داده ها
مشخص میکنیم در xکه کدام ستون شمرده شود و روی محور افقی رسم شود
در قسمتax=axبرای زمانی که چند نمودار داریم استفاده میکنیم ومیگوییم هر نمودار کجای صفحه در کدام خانه بکش
در ax.set_titleمشخص میکنیم اسم هر ستون بالای نمودار خودش نوشته شود
