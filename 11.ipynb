```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("heart.csv")
(df['target'].value_counts(normalize=True) * 100).plot(kind='bar')
plt.title("class Division")
plt.xlabel("class")
plt.ylabel("percentage")
plt.show()
```
