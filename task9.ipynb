اگر رکورد تکراری وجود دارد:

- چند عدد است؟
- چند درصد کل داده است؟

```python
df[df.duplicated()] 
```

با این متد تمام رکورد های تکراری هر ستون رو متوجه میشیم که بیشترین تکرار در ستون بیشرین ضربان قلب بوده

```python
df.duplicated().sum()  
```

با استفاده ازین دو متد فقط سطر هایی که تکراری هستند رئ متوجه میشیم که در اینحا فقط یکی از سطر ها تکراری است 

```python
(df.duplicated().sum() / len(df))
```

درصد تکرار رکورد ها خیلی پایینه در نتیجه داده های ما احتمالا کیفیت بالایی دارند و تاثیر بدی نمیزاره و نیازی به حذفشون نیست واحتمالا خطای ثبتی رخ نداده است
