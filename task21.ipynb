```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
df = pd.read_csv("data/heart.csv")
df
```

# بخش ششم: ارتباط ویژگی‌ها با هدف

### سوال 21

برای تمام ویژگی‌های عددی:

توزیع آنها را بر اساس target مقایسه کنید.

ابزار:

- boxplot
- violinplot

```python
fig, ax = plt.subplots(nrows=1 ,ncols=2 ,figsize= (15,5))
sns.violinplot(x="target",y="age",data=df,color = "skyblue" ,ax=ax[0])
ax[0].set_title("violin plot 'age'")
sns.violinplot(x="target",y="trestbps",data=df,color = "lightgreen",ax= ax[1])
ax[1].set_title("violin plot 'trestbps'")
plt.show()
```

کسانی که بیماری دارند و ندارند از روی سن قابل تشخیص نیستندولی کسانی که بیمار هستند پراکندگی سنشون بیشتره تا کسانی که بیمار نبودن و تست دادند 

فشارخون در افراد بیمار و افرادی ک بیمار نیستند تقریبا مشابه هستش و به صورت کلی با همین بدونه ستون نمیشه نتیجه خاصی گرفت

```python
fig, ax = plt.subplots(nrows=1 ,ncols=3 ,figsize= (35,10))
sns.boxplot(x="target",y="chol",data=df,color = "pink" ,ax=ax[0])
ax[0].set_title("violin plot 'chol'")
sns.boxplot(x="target",y="thalach",data=df,color = "lavender",ax= ax[1])
ax[1].set_title("violin plot 'thalach'")
sns.boxplot(x="target",y="oldpeak",data=df,color = "lightcoral",ax= ax[2])
ax[2].set_title("violin plot 'oldpeak'")
plt.show()
```

کسانی که بیماری قلبی دارند کلسترول تقریبا پایین تری داشتند و از این پارامتر نیز نمیشه شک کرد کی بیمار است

کسانی که بیمار هستند ضربان قلب بالاتری حین تمرین داشتند و اختلاف بین دو باکس بیمار و سالم زیاد هستش 

کسانی که بیماری قلبی دارند شیب اس تس خیلی کمتری نسبت افراد سالم دارند و از دو پارامتر شیب اس تس و ضحداکثر ضربان قلب بیشتر میشه شک به داشتن بیماری داشت
