```python
print("میانگین سن: ",df["age"].mean())
print("میانه سن: ",df["age"].median())
print("مد سن: ",df["age"].mode()[0])
```

باتوجه به تقریبا برابر بودن میانه ومیانگین میتوان گفت توزیع این ستون تقریبا متقارنه 

```python
print("میانگین فشارخون در حال استراحت :  ",df["trestbps"].mean())
print("میانه فشارخون در حال استراحت : ",df["trestbps"].median())
print("مد فشارخون در حال استراحت : ",df["trestbps"].mode()[0])
```

خیلی کم کشیدگی راست داریم ولی بیشتر انگار توزیع داده ها متقارنه 

```python
print("میانگین کلسترول خون :  ",df["chol"].mean())
print("میانه کلسترول خون  : ",df["chol"].median())
print("مد کلسترول خون: ",df["chol"].mode()[0])
```

کشیدگی چپ داریم یعنی توزیع داه ها به سمت مقادیر کمتر از میانگین هستش 

```python
print("میانگین درد ناشی از ورزش ",df["thalach"].mean())
print("میانه درد ناشی از ورزش",df["thalach"].median())
print("مد درد ناشی از ورزش",df["thalach"].mode()[0])
```

کشیدگی راست داریم یعنی بیشتر مقادیر از مانگین بیشتر هستند

```python
print(" میانگین شیب اس تس در تست ورزشی: ",df["oldpeak"].mean())
print("میانه شیب اس تس در تست ورزشی: " ,df["oldpeak"].median())
print("مد شیب اس تس در تست ورزشی: " ,df["oldpeak"].mode()[0])
```
