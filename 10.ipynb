Target نشان می دهد که آیا کاربر بیماری قلبی دارد یا ندارد همونطور که مشخصه 165نفر target مساوی 1 دارند که یعنی بیماری قلبی دارند و 138نفر target مساوی 0 که یعنی بیماری قلبی ندارند

```python
import seaborn as sns 
import matplotlib. pyplot as plt
sns. countplot (x="target", data=df)
plt. title( "Distribution of Target")
plt.show( )
```

در این نمودار می می توانیم فراوانی و توزیع دسته هارو تشخیص بدیم که همانطور که مشخصه تعداد افراد سالم که با 1 نشان داده شده بیشتره و در 0 که افراد بیمار رو نشون میده و توزیع نسبتاً نزدیک به هم هستش

```python
df["target"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    labels=["No Disease (0)", "Disease (1)"],
    startangle=90
)

plt.title("Target Distribution")
plt.ylabel("")
plt.show()
```

ما در این نمودار میبینیم که میزان افرادی که دارای بیماری قلبی هستند (Disease=1) و افرادی که بیماری قلبی ندارن (No Disease=0) که روی هم target  رو تشکیل میدن هر کدوم چند درصد از کل داده target هستند که میبینیم 45.5% رو افراد بیمار و 54.5% افراد سالم تشکیل دادن
