```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('data/heart.csv')


mean_age = df.groupby('target')['age'].mean()


plt.figure(figsize=(6,4))

mean_age.plot(kind='bar')

plt.xlabel('Target')
plt.ylabel('Average Age')
plt.title('Average Age by Heart Disease Class')

plt.xticks([0,1], ['No Disease', 'Disease'], rotation=0)

plt.show()
```
