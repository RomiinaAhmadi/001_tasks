```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
df = pd.read_csv("data/heart.csv")
df
```

### سوال 25

برای ویژگی oldpeak بررسی کنید:

رابطه آن با بیماری چیست؟

```python
df.groupby("target")["oldpeak"].mean()
```

کسانی که بیماری دارند شیب اس تی کمتری نسبت به کسانی که بیاری ندارند دارند

```python
sns.boxplot(x="target",y="oldpeak",data=df,color = "pink")
plt.show()
```

این نمودار نیز مانند جدول قبلی نشان دهنده این که کسایی که بیماری دارند شیب اس تی کمتری دارند


 چون مبانه و کلا جعبه پایین تر از کسایی هستش که بیماری ندارند
